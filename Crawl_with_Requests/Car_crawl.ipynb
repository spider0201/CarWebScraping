```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import csv

# Khởi tạo trình duyệt
driver = webdriver.Chrome()

# Hàm đóng thông báo cookie hoặc privacy
def close_popup():
    try:
        cookie_button = driver.find_element(By.ID, 'onetrust-accept-btn-handler')  # ID của nút Accept Cookies
        cookie_button.click()
        print("Đã đóng thông báo cookies.")
    except NoSuchElementException:
        print("Không tìm thấy thông báo cookies.")

# Hàm cuộn tới và nhấp vào nút Next
def click_next_button():
    try:
        next_button = driver.find_element(By.ID, 'next_paginate')
        driver.execute_script("arguments[0].scrollIntoView(true);", next_button)  # Cuộn tới nút
        time.sleep(1)
        next_button.click()
        print("Đã nhấp vào nút Next.")
    except NoSuchElementException:
        print("Không tìm thấy nút Next.")
    except Exception as e:
        print(f"Lỗi khi nhấp vào nút Next: {e}")

# Hàm crawl dữ liệu
def crawl_data():
    Title = []
    Stock_Status = []
    Price = []
    Dealer = []
    Rating = []
    Num_review = []

    page_number = 1

    while True:
        try:
            url = f'https://www.cars.com/shopping/results/?page={page_number}&page_size=100&zip=60606'
            driver.get(url)
            print(f"Đang crawl trang {page_number}...")

            close_popup()  # Đóng thông báo cookie trước khi crawl

            for _ in range(3):
                try:
                    titles = driver.find_elements(By.CSS_SELECTOR, 'a.vehicle-card-link h2.title')
                    Title.extend([title.text for title in titles])

                    stock_types = driver.find_elements(By.CSS_SELECTOR, 'p.stock-type')
                    Stock_Status.extend([stock.text for stock in stock_types])

                    prices = driver.find_elements(By.CSS_SELECTOR, 'div.price-section .primary-price')
                    Price.extend([price.text for price in prices])

                    dealers = driver.find_elements(By.CSS_SELECTOR, 'div.dealer-name strong')
                    Dealer.extend([dealer.text for dealer in dealers])

                    ratings = driver.find_elements(By.CSS_SELECTOR, 'spark-rating')
                    Rating.extend([rating.get_attribute("rating") for rating in ratings])

                    num_reviews = driver.find_elements(By.CSS_SELECTOR, 'spark-rating span')
                    Num_review.extend([num_review.text for num_review in num_reviews])

                    break
                except StaleElementReferenceException:
                    time.sleep(1)

            # Nhấp vào nút Next để chuyển trang
            click_next_button()
            time.sleep(2)  # Đợi trang tiếp theo tải
            page_number += 1

        except Exception as e:
            print(f"Lỗi khi crawl trang {page_number}: {e}")
            break

    save_to_csv(Title, Stock_Status, Price, Dealer, Rating, Num_review)

# Hàm lưu dữ liệu vào file CSV
def save_to_csv(Title, Stock_Status, Price, Dealer, Rating, Num_review):
    csv_file_path = 'vehicle_data.csv'

    with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)

        writer.writerow(['Title', 'Stock_Status', 'Price', 'Dealer', 'Rating', 'Num_review'])

        for row in zip(Title, Stock_Status, Price, Dealer, Rating, Num_review):
            writer.writerow(row)

    print(f"Data has been saved to {csv_file_path}")

# Gọi hàm crawl dữ liệu
crawl_data()

# Đóng trình duyệt sau khi crawl xong
driver.quit()
```
